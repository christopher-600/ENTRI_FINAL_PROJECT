```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\Admin\Desktop\PROJECT CHRIS\BMW sales data (2010-2024).csv"
bmw_data = pd.read_csv(file_path)
```

```python
print("Data Loaded Successfully!")
print(bmw_data.head())
print(bmw_data.info())
```

```python
sales_by_year = bmw_data.groupby('Year')['Sales_Volume'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=sales_by_year, x='Year', y='Sales_Volume', marker='o')
plt.title('Total BMW Sales Volume Over the Years')
plt.xlabel('Year')
plt.ylabel('Sales Volume')
plt.grid(True)
plt.show()
```

```python
#Top 10 BMW Models by Sales Volume
```

```python
sales_by_model = bmw_data.groupby('Model')['Sales_Volume'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sales_by_model.plot(kind='bar')
plt.title('Top 10 BMW Models by Sales Volume')
plt.xlabel('Model')
plt.ylabel('Sales Volume')
plt.xticks(rotation=45)
plt.show()
```

```python
#Market Share by Region
```

```python
sales_by_region = bmw_data.groupby('Region')['Sales_Volume'].sum()

plt.figure(figsize=(8,8))
sales_by_region.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Market Share by Region')
plt.ylabel('')
plt.show()
```

```python
#Yearly Sales Trend with Moving Average
```

```python
sales_by_year = bmw_data.groupby('Year')['Sales_Volume'].sum().reset_index()
sales_by_year['Moving_Avg'] = sales_by_year['Sales_Volume'].rolling(window=3).mean()

plt.figure(figsize=(12,6))
sns.lineplot(data=sales_by_year, x='Year', y='Sales_Volume', label='Annual Sales')
sns.lineplot(data=sales_by_year, x='Year', y='Moving_Avg', label='3-Year Moving Avg', linestyle='--')
plt.title('BMW Sales Trend with Moving Average')
plt.xlabel('Year')
plt.ylabel('Sales Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Regional Performance Over Time
```

```python
sales_region_year = bmw_data.groupby(['Year', 'Region'])['Sales_Volume'].sum().reset_index()

plt.figure(figsize=(12,8))
sns.lineplot(data=sales_region_year, x='Year', y='Sales_Volume', hue='Region')
plt.title('Regional Sales Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Sales Volume')
plt.legend(title='Region')
plt.grid(True)
plt.show()
```

```python
#Sales Classification vs Price and Mileage
```

```python
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
sns.boxplot(data=bmw_data, x='Sales_Classification', y='Price_USD')
plt.title('Price Distribution by Sales Classification')
plt.xlabel('Sales Classification')
plt.ylabel('Price (USD)')

plt.subplot(1, 2, 2)
sns.boxplot(data=bmw_data, x='Sales_Classification', y='Mileage_KM')
plt.title('Mileage Distribution by Sales Classification')
plt.xlabel('Sales Classification')
plt.ylabel('Mileage (KM)')

plt.tight_layout()
plt.show()
```
